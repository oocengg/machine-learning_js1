```python
# open the orange.txt
text = open('orange.txt', 'r').read()
```

```python
# import all of the packages
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# try to tokenizing the doc
rawtext = text.replace(':', '').replace('?', '').replace('!', '')

# split the sentence with (.) as a break to array
finaltext = rawtext.split('.')

#do the stopword removal in english
vectorizer = TfidfVectorizer(stop_words='english')

#do the feature extraxtion
response = vectorizer.fit_transform(finaltext)
print(response)
```

```python
words = vectorizer.get_feature_names()

print(words)
```

```python
df = pd.DataFrame(
    response.todense(),
    columns=vectorizer.get_feature_names()
)
df.transpose()
```
